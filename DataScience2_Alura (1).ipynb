```python
import pandas as pd
url = "https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv"
df = pd.read_csv(url)
```

```python
#Transformando meu index em país
df.set_index('País', inplace=True)
```

```python
#Criação de uma lista com os anos do dataFrame

anos = list(map(str, range(1980, 2014)))
```

```python
#Localizando e armazendo apenas os dados do Brasil da list
brasil = df.loc['Brasil', anos]
brasil
```

```python
#Criação de um dataFrame com esses dados
#Primeiro criamos um dicionario
#Tolist é usado para transformar em uma lista normal de py pq o dataFrame nn consegue interpretar os dados corretamente com uma series
brasil_dict = {"Ano": brasil.index.tolist(), "Imigrantes": brasil.values.tolist()}
#Transformando em dataFrame
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
#Para aumentar o tamanho do grafico de uma forma geral
#largura x altura
plt.figure(figsize=(8,4))
#grafico de linha- valor de x , valor de y
plt.plot(dados_brasil['Ano'], dados_brasil['Imigrantes'])
#Para mostrar apenas de 5 em 5 anos
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.ylabel("Total de Imigrantes")
plt.xlabel("Ano")
# plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.title("Evolução dos imigrantes do Brasil para o Canadá")
plt.show() #para tirar as informações de cima
```

```python
#DESAFIO SOZINHO
# Configura o estilo do Seaborn
sns.set_theme()
# Define a paleta de cores 'Dark2' para os gráficos
sns.set_palette('Dark2')

# Cria uma figura com 2 linhas e 2 colunas de subplots
fig, axs = plt.subplots(2, 2, figsize=(12,6))
# Ajusta o espaçamento entre os subplots e o espaço ao redor da figura
fig.subplots_adjust(hspace=0.5, wspace=0.3)
# Define o título geral da figura e ajusta a posição vertical usando o parâmetro 'y'
fig.suptitle('Imigração dos maiores países da América do Sul\npara o Canadá de 1980 a 2013', fontsize=20, y=1.05)

# Lista de países a serem plotados
list_americ = ["Brasil", "Colômbia", "Argentina", "Peru"]
# Filtra o DataFrame para incluir apenas os países e anos de interesse(anos está definido lá em cima)
df_pais = df.loc[list_americ, anos]


# Loop para criar 1 gráfico para cada país na lista
for i, ax in enumerate(axs.flat):
    pais = list_americ[i]

    # Criação dos graficos com sns
    #x=df_pais.columns: Define os anos como o eixo X.
    #y=df_pais.loc[pais]: Define os valores de imigração para o país específico como o eixo Y.
    #lw=3: Define a largura da linha para ser mais visível.
    #ax=ax: Plota o gráfico no subplot específico (ax) da figura
    sns.lineplot(x=df_pais.columns, y=df_pais.loc[pais], lw=3, ax=ax)

    # Define o localizador de ticks do eixo x para mostrar ticks a cada 5 unidades
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    # Define o título do subplot específico, alinhado à esquerda
    ax.set_title(f"País: {pais}", loc='left', fontsize=16)
    # Define o rótulo do eixo x e o tamanho da fonte
    ax.set_xlabel("Ano", fontsize=14)
    # Define o rótulo do eixo y e o tamanho da fonte
    ax.set_ylabel("Total de imigrantes", fontsize=14)

# Define os limites do eixo y para todos os subplots, mantendo um padrão
ymin = 0
ymax = 7000

# Aplica os limites definidos a todos os subplots
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

# Exibe a figura com todos os subplots
plt.show()
```

```python
import plotly.graph_objs as go  # Importa o módulo graph_objs da biblioteca Plotly para criar gráficos interativos

# Criando uma figura
fig = go.Figure()  # Cria um objeto Figure do Plotly, que é a base para o gráfico

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(  # Adiciona um trace (camada de dados) à figura, usando um objeto Scatter para um gráfico de dispersão
        x=[dados_brasil['Ano'].iloc[0]],  # Define o valor inicial do eixo X como o primeiro ano do DataFrame dados_brasil
        y=[dados_brasil['Imigrantes'].iloc[0]],  # Define o valor inicial do eixo Y como o primeiro valor da coluna 'Imigrantes'
        mode='lines',  # Define que o trace será exibido como uma linha
        name='Imigrantes',  # Define o nome do trace, que será usado na legenda do gráfico
        line=dict(width=4)  # Define a espessura da linha como 4 pixels
    )
)

# Definindo as configurações de layout
fig.update_layout(
    title=dict(  # Define o título do gráfico
        text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>',  # Define o texto do título, formatado em negrito
        x=0.12,  # Define a posição horizontal do título no layout
        xanchor='left',  # Define o alinhamento horizontal do título
        font=dict(size=20)  # Define o tamanho da fonte do título
    ),
    xaxis=dict(  # Define as configurações do eixo X
        range=[1980, 2013],  # Define o intervalo do eixo X
        autorange=False,  # Desabilita o ajuste automático do intervalo do eixo X
        title='<b>Ano</b>'  # Define o título do eixo X, formatado em negrito
    ),
    yaxis=dict(  # Define as configurações do eixo Y
        range=[0, 3000],  # Define o intervalo do eixo Y
        autorange=False,  # Desabilita o ajuste automático do intervalo do eixo Y
        title='<b>Número de imigrantes</b>'  # Define o título do eixo Y, formatado em negrito
    ),
    updatemenus=[dict(  # Define as configurações do menu de atualização
        type='buttons',  # Define que o menu será composto por botões
        showactive=False,  # Define que nenhum botão estará ativo inicialmente
        buttons=[dict(  # Define o botão que será exibido no menu
            label='Play',  # Define a etiqueta do botão
            method='animate',  # Define o método de atualização como 'animate' para ativar a animação
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]  # Define os argumentos para a animação
        )]
    )],
    width=1000,  # Define a largura do gráfico
    height=500  # Define a altura do gráfico
)

# Definir as configurações de animação
frames = [go.Frame(data=[go.Scatter(x=dados_brasil['Ano'].iloc[:i+1], y=dados_brasil['Imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))]  # Cria uma lista de frames para a animação
fig.frames = frames  # Atribui a lista de frames à propriedade 'frames' do objeto Figure

# Mostrando a figura
fig.show()  # Exibe o gráfico animado
```
